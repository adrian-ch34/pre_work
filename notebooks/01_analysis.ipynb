```python
# Step 0. Load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1. Load data
df_raw = pd.read_csv("../real_estate.csv", sep=";")
df_raw.sample(10)
```

```python
df_raw.info()
```

```python
# Step 2. Preprocessing  
df_baking = df_raw.copy()
df_baking = df_baking.iloc[:,1:-2]
df_baking["date"] = pd.to_datetime(df_baking["date"], errors="coerce")
df_baking.columns = df_baking.columns.str.lower()
df_baking = df_baking.drop(columns=["country","countryid","level1id","level2id","level3id","level4id","level5id","level6id","level7id","level8id","accuracy"])
df_baking["id_realestates"] = df_baking["id_realestates"].astype("category")
df_baking["phone_realestate"] = df_baking["phone_realestate"].astype("category")
df_baking["realestate_name"] = df_baking["realestate_name"].astype("category")
df_baking["url_inmueble"] = df_baking["url_inmueble"].astype("category")
df_baking["address"] = df_baking["address"].astype("category")
df_baking["description"] = df_baking["description"].astype("category")
df_baking["latitude"] = df_baking["latitude"].str.replace(",",".")
df_baking["latitude"] = df_baking["latitude"].astype(float)
df_baking["longitude"] = df_baking["longitude"].str.replace(",",".")
df_baking["longitude"] = df_baking["longitude"].astype(float)
df_baking.info()
```

```python
df = df_baking.copy()
df.sample(5,random_state=2025)
```

Print both average prices and then write a conclusion about them.


```python
# Filtramos precios por localidad
valdemorillo_prices = df[df["level5"] == 'Valdemorillo']['price']
galapagar_prices = df[df["level5"] == 'Galapagar']['price']

# Calculamos promedios
valdemorillo_avg = valdemorillo_prices.mean()
galapagar_avg = galapagar_prices.mean()

# Imprimimos los resultados
print(f"Precio promedio en Valdemorillo: {valdemorillo_avg:.2f}")
print(f"Precio promedio en Galapagar: {galapagar_avg:.2f}")

# Conclusión
if valdemorillo_avg == galapagar_avg:
    print("Conclusión: Los precios promedio en Valdemorillo y Galapagar son iguales.")
else:
    print("Conclusión: Los precios promedio en Valdemorillo y Galapagar son diferentes.")
```

Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in Exercise 14) (★★★)

Print both values for each variable.


```python
# Seleccionar solo las columnas de interés
variables = ["price", "rooms", "surface", "bathrooms"]

# Calcular la media muestral
sample_means = new_df[variables].mean()

# Calcular la varianza muestral (ddof=1 para que sea muestral)
sample_variances = new_df[variables].var(ddof=1)

# Imprimir los resultados
for var in variables:
    print(f"{var.capitalize()}:")
    print(f" Promedio Muestral = {sample_means[var]:.2f}")
    print(f" Varianza Muestral = {sample_variances[var]:.2f}")
    print()
```


Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the multihist demo of Matplotlib.


```python
# Copiar para no modificar el original
df_norm = new_df.copy()

# Aplicar normalización por población (level5)
df_norm["price_norm"] = df_norm.groupby("level5")["price"].transform(
    lambda x: (x - x.mean()) / x.std()
)

# Obtener las localidades únicas
locations = df_norm["level5"].unique()

# Crear histogramas
plt.figure(figsize=(10, 6))

# Graficar un histograma para cada localidad
for loc in locations:
    prices = df_norm[df_norm["level5"] == loc]["price_norm"]
    plt.hist(prices, bins=15, alpha=0.6, label=loc)

plt.title("Distribución normalizada de precio por localidad")
plt.xlabel("Precio normalizado")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Getafe muestra una distribución más ancha, con colas más largas a ambos lados, especialmente hacia la derecha (precios altos), lo que sugiere una mayor variabilidad de precios en esa zona.

Fuenlabrada, Alcorcón y Leganés tienen distribuciones más compactas y simétricas, lo que puede indicar mercados inmobiliarios más homogéneos.

Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)

Hint: Make a scatter plot of each population using subplots.


```python
# Obtener las 4 localidades únicas del subset
towns = new_df["level5"].unique()

# Crear subplots 2x2
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()  # Para iterar más fácilmente

# Crear un scatter plot por localidad
for i, town in enumerate(towns):
    town_data = new_df[new_df["level5"] == town]
    axes[i].scatter(town_data["surface"], town_data["price"], alpha=0.6)
    axes[i].set_title(f"{town}")
    axes[i].set_xlabel("Surface (m²)")
    axes[i].set_ylabel("Price (€)")
    axes[i].grid(True, linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```
